# Code sample using the Network Component library

```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
print(1)
from matplotlib.pyplot import *
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title')
show()
```
